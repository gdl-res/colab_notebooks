### What is Object Detection?

Object detection is a branch of computer vision that involves identifying and locating objects within an image or video. Unlike image classification, which only determines the presence of an object, object detection also provides the exact position of each object in the form of bounding boxes. This technology is widely used in various fields such as:
- **Autonomous Vehicles:** Detecting pedestrians, vehicles, and obstacles to ensure safe navigation.
- **Healthcare:** Identifying tumors in medical images or tracking patient activity.
- **Environmental Monitoring:** Monitoring natural indicators to help the conservation of habitats.
- **Security:** Real-time surveillance and threat detection.

Object detection can be applied to virtually anything, as long as the objects of interest are properly labeled in a dataset. By labeling images with categories and bounding boxes, models can be trained to detect custom objects, whether itâ€™s identifying plant species, detecting specific types of vehicles, or even finding defects in industrial products.

### What is Labelling?

Labeling is the process of annotating images or videos with information about the objects they contain. It's basically the way in which information is passed that is then used by the neural network to train.
In the case of object detection, this involves drawing bounding boxes around objects and assigning them appropriate categories, which serves as the foundation for training object detection models.

Tools like [Roboflow](https://roboflow.com/), [Label Studio](https://labelstud.io/), and datasets like [COCO](https://cocodataset.org/) provide powerful resources for labeling and dataset preparation. Blogs and documentation from these platforms, such as the [Roboflow Blog](https://blog.roboflow.com/) and [Label Studio Blog](https://labelstud.io/blog/), offer detailed guidance on creating high-quality labels for various applications.

### Data is the New Oil

Data is the backbone of object detection. High-quality, well-annotated datasets ensure that the models can perform accurately across diverse scenarios. The effectiveness of an object detection model is directly tied to the quantity and quality of the data it learns from. Key points about the importance of data are:

- **Variety:** A diverse dataset allows models to generalize well to new, unseen data.
- **Accuracy:** Precise labeling of objects ensures the model understands the nuances of the target classes.
- **Scalability:** The more comprehensive the dataset, the more adaptable the model becomes for real-world applications.
- **Bias Mitigation:** Balanced datasets prevent biases, ensuring fair and reliable predictions.

### Exploring Object Detection Models

There are many object detection models available today, each with unique features and use cases. Frameworks like [OpenMMLab](https://openmmlab.com/) or hubs like [Hugging Face](https://huggingface.co/) provide a collection of powerful models, including Faster R-CNN, Mask R-CNN, and Cascade R-CNN, which are widely used in research and industry.

Despite the variety of options, this notebook focuses on **YOLOv11**, a cutting-edge model known for its speed, simplicity, and performance. YOLO (You Only Look Once) is designed for real-time object detection, offering impressive accuracy with optimized computational efficiency. YOLOv11 builds on the advancements introduced in YOLOv9 and YOLOv10, incorporating improved architectural designs, enhanced feature extraction techniques, and optimized training methods.

### Why YOLOv11?

YOLOv11 is an excellent choice for object detection tasks due to:
- **Efficiency:** It achieves a higher mean Average Precision (mAP) score on the COCO dataset while using 22% fewer parameters than its predecessor, YOLOv8m, making it lightweight yet powerful.
- **Scalability:** Available in five different sizes, ranging from `2.6M` to `56.9M` parameters, YOLOv11 achieves between `39.5` and `54.7` mAP, allowing users to select a model variant that suits their needs.
- **Ease of Use:** YOLOv11's integration into frameworks like [Ultralytics](https://ultralytics.com/) provides simple APIs for training, testing, and deployment.

By choosing YOLOv11, we aim to balance cutting-edge performance with accessibility, making it good for both beginners and advanced practitioners in the field.

## CLI
CLI usage requires no customization or Python code. You can simply run all tasks from the terminal with the yolo command.

### Command Line Interface Overview
In this section, we'll explore the Command Line Interface (CLI) for YOLOv11, which offers a streamlined approach to executing various tasks without the need for custom Python scripts. By utilizing simple terminal commands, you can perform operations such as training, validation, prediction, and model export efficiently.

The general structure of a YOLOv11 CLI command is:
`yolo task=TASK mode=MODE [args]`

Where, according to the [documentation](https://docs.ultralytics.com/usage/cli/?utm_source=chatgpt.com):

- `TASK` specifies the type of task, such as `detect`, `segment`, `classify`, etc.

- `MODE` defines the operation mode, including `train`, `val`, `predict`, `export`, etc.

- `[args]` represents optional arguments that customize the command's behavior, like `model`, `data`, `epochs`, etc.

Here it is a [practical tutorial on the usage of CLI with YOLOv11](https://www.youtube.com/watch?v=bjkt5OE_ANA)

### Inference using CLI example

For an example of how to perform inference exploiting the CLI and a pretrained on COCO model, try running the following command:

```python
!yolo task=detect mode=predict model=yolo11n.pt conf=0.25 source='https://tenlives.com.au/wp-content/uploads/2020/09/Problem-Too-Many-Cats-scaled.jpg' save=True
```

**Command overview:**

- `yolo`: Executes the YOLO command-line interface (CLI) within a Jupyter Notebook cell.

- `task=detect`: Specifies the task as object detection.

- `mode=predict`: Sets the mode to prediction (inference) on new data.

- `model=yolo11n.pt`: Indicates the use of the YOLOv11 nano model for inference.

- `conf=0.25`: Sets the confidence threshold to 25%; detections with confidence below this value will be disregarded.

- `source='https://media.roboflow.com/notebooks/examples/dog.jpeg'`: Specifies the input image URL for object detection.

- `save=True`: Ensures that the output image, with detected objects annotated, is saved.

Result annotated image got saved in `{HOME}/runs/detect/predict/`. We can use the following code to display it:
